```python
def plu(A):
    
    ####################################################
    # Factorizes a square matrix A as PLU = A where    #
    # L is a lower triangular matrix, U is an upper    #
    # triangular matrix, and P is a permutation        #
    # matrix which keeps track of partial pivoting.    #
    ####################################################
     
    m = A.shape[0]
    
    P = np.eye(m)
    L = np.zeros((m, m))
    U = A.copy()
    
    # For each row, except the last row
    # which has no pivot column beneath its pivot.
    for r in np.arange(0, m - 1):
        
        # Get the pivot_col under the pivot
        # in r and find the index of the max
        # absolute value, col_index, in pivot_col.
        # Then, add r to col_index to return to
        # indexing of U.
        pivot_col = U[r:, r]
        col_index = np.argmax(abs(pivot_col))
        index = col_index + r 
        
        # If the max absolute value is
        # not in row r, swap row r with
        # row index; perform the same
        # permutations on P and L.
        if index != r:
            
            # Swap rows.
            P[[r, index], :] = P[[index, r], :]
            L[[r, index], :] = L[[index, r], :] 
            U[[r, index], :] = U[[index, r], :] 
          
        # Initialize matrix of scalars which when
        # applied to U makes all elements
        # below pivot row r 0.
        scalar_matrix = np.eye(m)
        
        # For each row s below the current row r,
        # calculate the scalars of the row 
        # operations necessary 
        # to make all elements 0.
        for s in np.arange(r + 1, m):
            scalar = U[s, r] / U[r, r]
            scalar_matrix[s, r]  = -scalar
            L[s, r] =  scalar
        
        # Make all elements below r 0.
        U = np.dot(scalar_matrix, U)
        
    # Fill diagonal of L with 1s to
    # yield a unique decomposition
    # into upper triangular and 
    # unit lower triangular matrices.
    np.fill_diagonal(L, 1)
    
    return P, L, U
```

```python
def qr_householder(A):
    
    ##############################################
    # Factorizes an m x n matrix A as QR = A     #    
    # where Q is an orthogonal matrix and        #
    # R is an upper triangular matrix            #
    # using Householder reflections.             #
    ##############################################
    
    m, n = A.shape
    
    R = A.copy()
    Q = np.eye(m)
    
    # For each col index c in R = A,
    # except the last column.
    for c in np.arange(0, n):
        
        # Get column vector at c.
        # Note that in order to be able
        # to take the transpose of col
        # later, we make the
        # column vector into a
        # 2D matrix; recall that the
        # transpose of a vector is that
        # same vector.
        col = R[c:, c, np.newaxis]
        
        # Initialize Householder 
        # matrix for vector col.
        H = np.eye(m)
        
        # Populate Householder matrix for col.
        v = col / (col[0] + np.copysign(scipy.linalg.norm(col), col[0]))
        # Note that v[0] = 1
        # is the equivalent of
        # multiplying v by the
        # first standard basis
        # vector e_1 = [1, 0, 0, ...]^T.
        v[0] = 1
        col_m = col.shape[0]
        H[c:, c:] = np.eye(col_m) - (2 / np.dot(v.T, v)) * np.dot(v, v.T)
        
        # Note that we swap the order
        # of the product Q = HQ
        # in order to avoid an additional 
        # Q.T before outputting.
        Q = np.dot(Q, H)
        R = np.dot(H, R)
        
    return Q, R
```

```python
def qr_cgs(A):
    
    m, n = A.shape
    
    Q = np.zeros((m, n))
    R = np.zeros((n, n))

    for j in np.arange(0, n):
        
        v = A[:, j]

        for i in np.arange(0, j):

            q = Q[:, i]
            R[i, j] = np.dot(q, A[:, j])
            v = v - R[i, j] * q

        norm = scipy.linalg.norm(v)
        R[j, j] = norm
        Q[:, j] = v / norm
        
    return Q, R

def qr_mgs(A):
    
    m, n = A.shape
    
    Q = np.zeros((m, n))
    R = np.zeros((n, n))

    for j in np.arange(0, n):
        
        v = A[:, j]

        for i in np.arange(0, j):
            
            q = Q[:, i]
            R[i, j] = np.dot(q, v)  # Only difference from CGS.
            v = v - R[i, j] * q

        norm = scipy.linalg.norm(v)
        R[j, j] = norm
        Q[:, j] = v / norm
        
    return Q, R
```

```python
A = scipy.linalg.hilbert(6)
b = np.arange(0, 6)

def forward_substitution(L, b):
    
    ###########################################
    # Solves the system Ly = b for y, where L #
    # is a lower triangular matrix L.         #
    ###########################################
    
    # Get number of rows in L.
    m = L.shape[0]
    
    # Instantiate solution vector y
    # with shape of b. 
    y = np.zeros_like(b, dtype = np.double)
    
    y[0] = b[0] / L[0, 0]
    
    for i in range(1, m):
        
        y[i] = (b[i] - np.dot(L[i, :i], y[:i])) / L[i, i]
        
    return y

def back_substitution(U, y):
    
    ###########################################
    # Solves the system Ux = y for x, where U #
    # is an upper triangular matrix.          #
    ###########################################
    
    # Get number of rows in U.
    m = U.shape[0]
    
    # Instantiate solution vector x
    # with shape of y.
    x = np.zeros_like(y, dtype = np.double)

    x[-1] = y[-1] / U[-1, -1]
    
    # Looping through rows backwards,
    # omitting the second to last row 
    # since we already set it above.
    for i in np.arange(m-2, -1, -1):
        
        x[i] = (y[i] - np.dot(U[i, i:], x[i:])) / U[i, i]
                      
    return x

def solve_by_decomposition(A, b, method = "plu"):
    
    if method == "plu":
        
        ###########################################
        # Solves the system Ax = b through an     #
        # LU decomposition (w/ partial pivoting)  #
        # method.                                 #
        ###########################################

        P, L, U = plu(A)

        y = forward_substitution(L, np.dot(P, b))

        return back_substitution(U, y)
    
    if method == "qr-h":
        
        ###########################################
        # Solves the system Ax = b through a QR   #
        # decomposition found using Householder   #
        # reflections.                            #
        ###########################################

        Q, R = qr_householder(A)

        y = np.dot(Q.T, b)

        return back_substitution(R, y)
    
    if method == "qr-cgs":
        
        ###########################################
        # Solves the system Ax = b through a QR   #
        # decomposition found using a modified    #
        # Gram-Schmidt (MGS) method.              #
        ###########################################

        Q, R = qr_cgs(A)

        y = np.dot(Q.T, b)

        return back_substitution(R, y)
    
    if method == "qr-mgs":
        
        ###########################################
        # Solves the system Ax = b through a QR   #
        # decomposition found using a modified    #
        # Gram-Schmidt (MGS) method.              #
        ###########################################

        Q, R = qr_mgs(A)

        y = np.dot(Q.T, b)

        return back_substitution(R, y)
print(qr_householder_test(A))
Q, R = qr_householder(A)
print(R)
print(f"PLU method:\n\n{(solve_by_decomposition(A, b, 'plu'))}\n--------------------------------------")
print(f"QR Householder method:\n\n{solve_by_decomposition(A, b, 'qr-h')}\n--------------------------------------")
print(f"QR CGS method:\n\n{solve_by_decomposition(A, b, 'qr-cgs')}\n--------------------------------------")
print(f"QR MGS method:\n\n{solve_by_decomposition(A, b, 'qr-mgs')}\n--------------------------------------")
print(f"Scipy (PLU) method:\n\n{scipy.linalg.solve(A, b)}\n--------------------------------------")
```

```python
def time_benchmark(m, n_array, method):
    
    ###########################################################################
    # Solves the system Ax = b for m n x n systems with the provided method   #
    # where n is taken from n_array each iteration and is typically           #
    # an increasing sequence (eg. n_array = np.arange(10, 210, 10) takes the  #
    # average of 10 runs for each of n = (10, 20, 30, 40, ..., 200)).         #
    #                                                                         #
    # The method is timed for each iteration and the mean times to            #
    # solve an n x n system outputted as an array.                            #
    ###########################################################################
    
    avg_t_array = np.array([])
    
    # For each matrix size n in n_array.
    for n in n_array:
        
        # Create empty array to store elapsed time values t.
        t_array = np.array([])
        
        # Solve m different matrices of size n
        # and time it.
        for i in np.arange(0, m):
            
            A = np.random.rand(n, n)
            b = np.random.rand(n)

            start = time.perf_counter()
            solve_by_decomposition(A, b, method)
            t = (time.perf_counter() - start)

            t_array = np.append(t_array, t)
        
        # Get mean time elapsed to solve n x n system.
        avg_t_array = np.append(avg_t_array, np.mean(t_array))
    
    return avg_t_array

def error_benchmark(m, n_array, method):
    
    ###########################################################################
    # Solves the system Ax = b for m n x n systems with the provided method   #
    # where n is taken from n_array each iteration and is typically           #
    # an increasing sequence [eg. n_array = np.arange(10, 210, 10) takes the  #
    # average of 10 runs for each of n = (10, 20, 30, 40, ..., 200)].         #
    #                                                                         #
    # The method's error is evaluated each iteration and the median 2-norm    #
    # error in solving an n x n system outputted as an array.                 #
    ###########################################################################
    
    avg_err_array = np.array([])
    
    # For each matrix size n in n_array.
    for n in n_array:
        
        # Create empty array to store errors err.
        err_array = np.array([])
        
        # Solve m different matrices of size n
        # and get the mean absolute error
        # between Ax and b.
        for i in np.arange(0, m):
            
            A = np.random.rand(n, n)
            b = np.random.rand(n)
        
            # Get solution using method.
            x = solve_by_decomposition(A, b, method)
            
            # Calculate 2-norm error.
            err = scipy.linalg.norm(b - np.dot(A, x))
            
            err_array = np.append(err_array, err)
        
        # Get median 2-norm error in solving n x n system.
        avg_err_array = np.append(avg_err_array, np.median(err_array))
    
    return avg_err_array
```

```python
n_min_t = 5
n_max_t = 250
n_step_t = 5
m_t = 10

t_plu = time_benchmark(m_t, np.arange(n_min_t, n_max_t, n_step_t), "plu")
t_qr_h = time_benchmark(m_t, np.arange(n_min_t, n_max_t, n_step_t), "qr-h")
t_qr_cgs = time_benchmark(m_t, np.arange(n_min_t, n_max_t, n_step_t), "qr-cgs")
t_qr_mgs = time_benchmark(m_t, np.arange(n_min_t, n_max_t, n_step_t), "qr-mgs")
```

```python
#plt.scatter(x, y, color = "black", s = 1, zorder = 5)

fig = plt.figure(constrained_layout = True, figsize = (6, 4))
ax = plt.gca()

plt.plot(np.linspace(n_min_t, n_max_t, len(t_plu)), t_plu, label = "PLU", color = "lightsteelblue")
plt.plot(np.linspace(n_min_t, n_max_t, len(t_qr_h)), t_qr_h, label = "QR (Householder)", color = "darkseagreen")
plt.plot(np.linspace(n_min_t, n_max_t, len(t_qr_mgs)), t_qr_mgs, label = "QR (MGS)", color = "lightcoral")
plt.plot(np.linspace(n_min_t, n_max_t, len(t_qr_cgs)), t_qr_cgs, label = "QR (CGS)", color = "mediumpurple")

plt.grid(linestyle = "--")
#plt.grid(which = "major", linestyle = "--", color = "black")
plt.xlabel("Input size")
plt.ylabel("Time (s)")

plt.xlim(0)
plt.ylim(0)

plt.legend(framealpha = 1)

ax.minorticks_on()

plt.savefig("Time250.pdf", bbox_inches = "tight")
```
